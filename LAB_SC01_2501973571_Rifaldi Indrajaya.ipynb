Intro IF Else

```python
def myThermoStat(temp, desTemp):
    if temp < desTemp - 5:
      status = 'Heat'
    elif temp > desTemp + 5:
      status = 'AC'
    else:
      status = 'OFF'
    return status
```

```python
status = myThermoStat(65,75)
print(status)
```

```python
status = myThermoStat(75,65)
print(status)
```

```python
status = myThermoStat(65,63)
print(status)
```

Mencoba Fungsi

```python
##function sum matrix two dimentional

def sum_matrix(matrix):
    sum = 0
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            sum += matrix[i][j]
    return sum

##funtion bubble sort

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

```python
sum_matrix([[1,2,3],[4,5,6]])
```

```python
bubble_sort([10,2,12,4,87,6])
```
